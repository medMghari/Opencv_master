## Jupyter Platforme 

Nous utiliserons l'interface du Jupyter notebook pour tous nos travaux.
Jupyter Notebook est une application Web open source qui vous permet de créer et de partager des documents contenant du code en direct, des équations, des visualisations et du texte narratif. Les utilisations incluent le nettoyage et la transformation des données, la simulation numérique, la modélisation statistique, la visualisation des données, l'apprentissage automatique et bien plus encore.
Le notebook est divisé en cellules, dont certaines sont des `markdowns`.

```python
print('cette cellule contient du code')
```

Le code que vous ecrivez dans le naviguateur est envoyé au noyau Python qui est localisé; soit dans votre propre machine si vous travailler avec `Anaconda`, soit dans le cloud si vous avez choisi `google colab`. 
Dans les deux cas le serveur va executer votre code et retourn un resultat au notebook.
Pour executer le cellule vous clickez sur le boutton à gauche de le cellule, ou bien vous tappez `shift+enter`.

## Installation

#### Google Colab
Si vous avez une compte Gmail donc tout est déja fait pour vous. il suffit d'aller à votre  Google Drive en clickant “Nouveau” pour créer un nouveau dossier et saisir le nom suivant "Colab notebooks" (ce nom n'est obligatoire c'est juste une convention).

Aprés avoir créer le dossier, vous pouvez créer un nouveau notebook Colab. Cliquez simplement sur "Nouveau" et faites glisser le menu jusqu'à "Plus", puis sélectionnez "Colaboratory".
