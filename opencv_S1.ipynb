Avec OpenCV

Que s'est-il passé? <br>
Pour des raisons historiques, OpenCV utilise par défaut le format BGR au lieu du RGB habituel<br><br>
Solutions: <br>
*   Utilisez numpy/OpenCV de manière cohérente<br>
*   Ou assurez-vous de convertir<br>

````python
cvimrgb = cv2.cvtColor(cvim2disp,cv2.COLOR_BGR2RGB)

#Ou bien
#imbgr = cv2.cvtColor(im2disp,cv2.COLOR_RGB2BGR)

figure()
imshow(cvimrgb)
````

### Une fonction pour l'affichage

````python
def mycvshow(imagein, title='Image'): 
    cv2.imshow(title,imagein)
    cv2.waitKey()
    cv2.destroyWindow(title)

#une fonction qu'on va utiliser ulterieurement
````

Que s'est-il passé? Pourquoi la modification de y a-t-elle changé x ?
